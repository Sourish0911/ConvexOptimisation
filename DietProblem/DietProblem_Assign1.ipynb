```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
n = 3
cal_B = np.matrix([72,121,65])
Vit_A = np.matrix([107,500,0])
cost = np.matrix([0.18,0.23,0.05])
s = cp.Variable(n)

Constraints = [10 >= s, 0 <= s, 50000 >= cp.matmul(Vit_A,s) , 5000 <= cp.matmul(Vit_A,s), 2250 >= cp.matmul(cal_B,s) , 2000 <= cp.matmul(cal_B,s)]

Objective = cp.Minimize(cp.matmul(cost,s))

prob = cp.Problem(Objective, Constraints)
prob.solve()


print("Optimal Serv:",s.value)
print("Optimal Cost:", (cp.matmul(cost,s)).value)
```
