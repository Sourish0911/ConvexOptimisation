```python
import numpy as np
import cvxpy as cp
import pandas as pd
days=5


A=df.to_numpy()

x=cp.Variable((3,1))

Produced_per_day = cp.multiply(np.reshape(A[:,0],(len(A[:,0]),1)),x)
Produced_per_week = days*Produced_per_day


Objective= cp.Maximize(A[:,4].T@Produced_per_week)

#Constraints
proportion=[sum(x)<=1]
positive=[x>=0]
storage=[sum(cp.multiply(Produced_per_week , np.reshape(A[:,1],(3,1))/1000)) <= 6000]
demand=[ Produced_per_week <= np.reshape(A[:,3],(3,1)) ]
minSale=[Produced_per_week[0:2,:]>=np.reshape(A[0:2,2],(2,1))]



constraints=   storage  + demand + positive +proportion + minSale

prob = cp.Problem(Objective, constraints) 
prob.solve()
print( "Value of net profit is", prob.solve())
print("proportion values are", x.value)



```
